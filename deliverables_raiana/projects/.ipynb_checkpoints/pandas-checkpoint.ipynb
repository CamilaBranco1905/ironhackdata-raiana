```python
#importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
# reading attack.csv files as a pandas dataframe... the encoding is not the default ufc=8 it is latin-1
attack = pd.read_csv('attacks.csv',encoding = "latin-1")

```
