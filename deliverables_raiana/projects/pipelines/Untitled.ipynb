```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
suicide = pd.read_csv("suicide-rates-overview-1985-to-2016.zip")
```

```python
suicide.head(10)
country = suicide.groupby("country")
```

```python
#what do i wanna know? brazil general tendency, gender diff, age diff compare with global
```

```python
brsui = country.get_group("Brazil")
brsui.head(10)
```

```python
year = brsui.groupby("year")
years = year.sum()
```

```python
plt.plot([years.index], [years["suicides/100k pop"]])
```
