```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
suicide = pd.read_csv("suicide-rates-overview-1985-to-2016.zip")
homicide = pd.read_csv("brazilian-number-of-murders-1996-to-2016-ipea.csv",sep=';')
homicide2 = pd.read_csv("homicide-reports.zip")
```

```python
suicide.head()
```

```python
scountry = suicide.groupby("country")
sbr = scountry.get_group("Brazil").groupby("year").sum()
susa = scountry.get_group("United States").groupby("year").sum()
```

```python
hbrazil = homicide.groupby("período").sum()
hbrazil = hbrazil.rename({"período": "Year"}).drop(["cod",axis=1)
```

```python
homicide2.loc[homicide2["Victim Count"]==0,"Victim Count"]=1
hamerica = homicide2["Victim Count"].groupby(homicide2["Year"]).sum()
```

```python
##dbr1.loc["suicides_no"/("population"/100000)] = "suic/100k"
```

```python
dbr["suic/100k"] = dbr.apply(lambda row: row.suicides_no/(row.population/100000), axis=1)
dbr["homi/100k"] = dbr.apply(lambda row: row.valor/(row.population/100000), axis=1)
dusa["suic/100k"] = dusa.apply(lambda row: row.suicides_no/(row.population/100000), axis=1)
dusa["homi/100k"] = dusa.apply(lambda row: row.valor/(row.population/100000), axis=1)
dc = dbr.join(dusa,lsuffix='_br', rsuffix='_usa')
```

```python
sns.lineplot(data=dc[["suic/100k_br","homi/100k_br","suic/100k_usa","homi/100k_usa"]], palette="rainbow", linewidth=2.5)
```
